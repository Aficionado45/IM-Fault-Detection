```python
import pandas as pd
```

```python
data = pd.read_csv(r'D:\Acads\BTP\Lightly Loaded\Correlation\10_15_BF_heat.csv')
```

```python
data.head()
data2=data.head(100)
```

```python
import seaborn as sns
```

```python
corr = data.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
